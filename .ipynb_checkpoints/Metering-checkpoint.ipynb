```python
# Create a unique DataFrame based on 'Work Order Name' to assign Subgroup once
unique_work_orders = df[['Work Order Name']].drop_duplicates()

# Create a function to assign 'Subgroup' based on conditions
def assign_subgroup(work_order_name):
    if work_order_name in work_order_names_moveIN:
        return 'Move-In Kit'
    elif work_order_name in work_order_names_Apt:
        return 'Apartment Kit'
    else:
        return 'HEC'

# Apply the function to the unique DataFrame
unique_work_orders['Subgroup'] = unique_work_orders['Work Order Name'].apply(assign_subgroup)

# Now map the subgroup back to the original dataframe
df['Subgroup'] = df['Work Order Name'].map(unique_work_orders.set_index('Work Order Name')['Subgroup'])

display_scrollable_dataframe(df)
```

```python
import numpy as np

def identify_outliers_zscore(df, threshold=3):
    outliers = pd.DataFrame(index=df.index)
    for col in df.select_dtypes(include=[np.number]).columns:
        z_scores = (df[col] - df[col].mean()) / df[col].std()
        outliers[col] = np.abs(z_scores) > threshold
    return outliers

# Identify outliers
outliers_zscore = identify_outliers_zscore(df)

# Filter to show only rows with at least one True value
rows_with_outliers = outliers_zscore.any(axis=1)
outliers_only = df[rows_with_outliers]

# Count the number of outliers by field
outliers_count = outliers_zscore.sum()

print("\nNumber of outliers by field:")
print(outliers_count)
print(f"\nTotal outlier count: {outliers_count.sum()}")
print(f"Unique outlier count: {len(rows_with_outliers[rows_with_outliers == True])}")
```

### Locations

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

# Capture the start time
start_time = time.time()

# Initialize the geocoder (using OpenStreetMap's Nominatim)
geolocator = Nominatim(user_agent="geoapi")

# Assuming df_HEC is your dataframe containing a 'Site Address' column
# addresses = df_HEC['Site Address'].unique()
addresses = ["548 Desert Maize Dr SW, Albuquerque, NM, USA 87121"]

# Lists to store the latitude and longitude results
latitude = []
longitude = []

# Iterate over the addresses
for address in addresses:
    try:
        # Geocode the address
        location = geolocator.geocode(address)
        
        if location:
            # If geocoding is successful, append latitude and longitude
            latitude.append(location.latitude)
            longitude.append(location.longitude)
        else:
            # If address not found, append None
            latitude.append('No Match')
            longitude.append('No Match')
        
        # Pause for 1 second to avoid overwhelming the geocoding service
        time.sleep(0.1)
    
    except Exception as e:
        print(f"Error geocoding {address}: {e}")
        latitude.append(None)
        longitude.append(None)

# Create a DataFrame from the lists
df_location = pd.DataFrame({
    'Address': addresses,
    'Latitude': latitude,
    'Longitude': longitude
})

print(len(addresses))
display_scrollable_dataframe(df_location[df_location['Latitude'] != None],max_rows=1207)

# # Display the updated dataframe with latitude and longitude
# print(df_HEC.head())

# Capture the end time
end_time = time.time()

# Calculate the elapsed time in seconds
elapsed_time = end_time - start_time

print(f"Code execution time: {elapsed_time:.2f} seconds")
```

## Sample File Creation

```python
# Group by Project ID and aggregate numeric columns
aggregated_df = df.groupby(project_id_field)[numeric_columns].sum().reset_index()
aggregated_df = aggregated_df.set_index(project_id_field)
aggregated_df = aggregated_df.drop(columns=['Gkwh_perGkw','Gkwh_perIncen','ObservationID'])

# Group by 'Project ID' and get the first non-null 'Subgroup' value to include Subgroup in aggregated df
# Assign the 'Subgroup' value to 'SampleCategory_Primary' in the aggregated DataFrame
project_subgroup = df.groupby(project_id_field)['Subgroup'].first()
aggregated_df['SampleCategory_Primary'] = aggregated_df.index.map(project_subgroup)

# Print and display the total count and df of the newly aggregated df
print(f"Total Project Count: {len(aggregated_df):.0f}")
display_scrollable_dataframe(aggregated_df)

# Create a dictionary with Project ID as keys and lists of ObservationIDs as values
project_measure_dict = df.groupby(project_id_field)[observation_id_field].apply(list).to_dict()

# Convert the dictionary to a DataFrame for exporting to CSV
proj_measure_dict_df = pd.DataFrame(list(project_measure_dict.items()), columns=[project_id_field, observation_id_field])
proj_measure_dict_df = proj_measure_dict_df.set_index(project_id_field)
display_scrollable_dataframe(proj_measure_dict_df)
```

```python
from datetime import datetime

# Assign Program Name
prog_name = 'ResComp'

#Add DI designation for sampling group
aggregated_df['SampleGroup'] = 'HEC'

# Get the current date and format it as YYYY_MM_DD
current_date = datetime.now().strftime("%Y_%m_%d")

# Create the filename with the dynamic date
file_name = f"{prog_name}_{aggregated_df['SampleGroup'].iloc[0]}_{current_date}.xlsx"

# Write DataFrame to Excel
with pd.ExcelWriter(file_name) as writer:
    df.to_excel(writer, sheet_name='RawData', index=True)
    aggregated_df.to_excel(writer, sheet_name='PopByProject', index=True)
    proj_measure_dict_df.to_excel(writer, sheet_name='MeasToProj_Dict', index=True)
```
